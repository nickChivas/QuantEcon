```python
import numpy as np
import matplotlib.pyplot as plt

'''
In this application, we plot the balance of a bank account over time.

There are no withdraws over the time period, the last date of which is denoted by T. The initial balance is b_0 and the interest rate is r.
The balance updates from period t to t + 1 according to b_{t+1} = (1 + r)b_t. In the code below, we generate and plot the sequence b_0, b_1, ..., b_T.

Instead of using a Python list to store this sequence, we will use a NumPy array.
'''

r = 0.025           # interest rate
T = 50              # end date
b = np.empty(T+1)   # empty numPy array, stores all b_t. Allocates storage in memory for T+1 (floating point) numbers.
b[0] = 10           # initial balance

for t in range(T):
    b[t+1] = (1 + r) * b[t] # Updates the value of the index after the current by multiplying the current value by 1 plus the fixed interest rate.

plt.plot(b, label='bank balance') # Plots all the points in the array, labeled bank balance.
plt.legend() # Adds a legend to the pyplot.
plt.show()
```
