```python
import numpy as np # Importing numpy, a Python package that works with arrays, mathematical functions, number generating, linear algebra, etc. Access the functions using np.().
import matplotlib.pyplot as plt # Importing matplotlib, a package that is useful for data visualization.

e_values = np.random.randn(100) # Assigns an array of size 100 filled with values using the randn() method of the random subpackage from numpy
plt.plot(e_values) # Method from pyplot which allows me to plot the points from the e_values array.
plt.show() # Shows the plot in the notebook.
```

```python
ts_length = 100 # Desired length of the time series
e_values = [] # empty array

for i in range(ts_length): # Loop which assigns a random integer e and appends it to the array
    e = np.random.randn()
    e_values.append(e)

plt.plot(e_values)
plt.show()
```

```python
ts_length = 100
e_values = []
i = 0
while i < ts_length:
    e = np.random.randn()
    e_values.append(e)
    i += 1
plt.plot(e_values)
plt.show()
```
